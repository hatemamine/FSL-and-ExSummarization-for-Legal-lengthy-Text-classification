```python
import pandas
import numpy as np
df = pandas.read_csv("../input/output-ledgarlexglue-preprocesse-generate-dataset/LEDGARvalidation.csv")
df0 = pandas.read_csv("../input/output-ledgarlexglue-preprocesse-generate-dataset/LEDGARtrain.csv")
print(len(df))
print(len(df0))
#print(df.head())
```

```python
from tqdm.notebook import tqdm
import random
import numpy
random.seed(1)
embeddings=[]
lableskey=[]
import time
# get the start time
st = time.time()
for j in tqdm(range(len(df))):
    row =df.iloc[j]
    lableskey.append(row["MotsCles"])
    embeddings.append(model.encode(row["ATexte"], convert_to_tensor=True).cpu().numpy())
    #break
# get the end time
et = time.time()
# get the execution time
elapsed_time = et - st
print('Execution time:', elapsed_time, 'seconds')
import pickle
embeddings0 = {'embeddings_validation':embeddings, 'lable':lableskey}
with open('embeddingsledgar.pickle', 'wb') as handle:
    pickle.dump(embeddings0, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
